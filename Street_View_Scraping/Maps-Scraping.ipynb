```python
############### Scrape all google street view images in an area ################

#This module downloads all google maps images within a rectangular area
#This consists of two functions
#1. Generate a list of photo ID's within the bounds of four lat/long coordinates
#2. Download a set of photos from a list of photo ID's


#https://github.com/robolyst/streetview
#this repo can be used to find the closest panoramas to your current loc


######### Imports #############

import streetview





########## Generate a list of all photo ID's in your precinct #########



def getIdsInRectangle(SP,N_Border,S_Border,W_Border,E_Border):
    #check input is correct
    if N_Border < S_Border:
        raise InputError("Northern boundary must be greater lat than south")
    if E_Border < W_Border:
        raise InputError("Eastern Boundary must be greater than west")
    
    
    #init lists
    not_checked_IDs = []
    precinct_IDs = []

    #init position top left corner
    position=SP

    #get initial set of id's from coordinates
    found_panoids = streetview.panoids(position[0], position[1])

    new = 0
    iter_num = 0
    while True:
        iter_num+=1
        #if iter_num > 50:
        #    break
        if (iter_num%50==0):
            print("Iteration number is",iter_num,"with",len(not_checked_IDs),"Number of unchecked images left")
        #for the current pano, get all new nearby panoids
        #print(len(found_panoids), "Images being checked in this batch")
        for i in range(len(found_panoids)):

            pano_id = found_panoids[i]['panoid']
            position = (found_panoids[i]['lat'],found_panoids[1]['lon'])

            #store id if within boundaries and not already added
            if (((position[0] < N_Border) and (position[0] >S_Border)) \
                and ((position[1] > W_Border) and (position[1] < E_Border))):
                #print(pano_id)
                #print(precinct_IDs)
                if pano_id not in precinct_IDs:
                    #print("addding pano")
                    precinct_IDs.append(pano_id)
                    not_checked_IDs.append({'panoid':pano_id,'lat':position[0],'lon':position[1]})
                    new+=1

        #print(new,"New images added in this iter")
        #check to see if any new images found
        if (len(not_checked_IDs)>0):
            #get a new image and new set of panoids if more to be checked
            new_pano = not_checked_IDs.pop(0)
            found_panoids = streetview.panoids(new_pano['lat'], new_pano['lon'])
            #print(len(found_panoids), "Images in this batch")
            new=0
        else:
            #if no new images then end loop
            break
    print(len(precinct_IDs),"Total ids found in the area")

    return precinct_IDs



    ######### Download a list of photo ID's into a folder ############





#streetViewPanoramaOptions
#position = {lat: -33.8291, lng: 151.2441}
#pov = {heading: 100, pitch: 0}
#zoom = 1

#Testing
#panoids = streetview.panoids(lat=-33.85693857571269, lon=151.2144895142714)
#print(panoids)


#print(panoids[0]['panoid'])
#print(panoids[0])
#print(panoids[1])
#https://maps.googleapis.com/maps/api/streetview/metadata?size=630x630&pano=pV6jtRc157XZtWpVIR-rtg&fov=90&heading=180&pitch=10&key=AIzaSyBCNskTKxgdbmwCh4BpVH0oo5-Xqt87MvY
#streetview.api_download(panoids[0]['panoid'], 180, "test_dir", int(googleMapsApiKey))
#https://maps.googleapis.com/maps/api/streetview?size=630x630&pano=pV6jtRc157XZtWpVIR-rtg&fov=90&heading=180&pitch=10&key=AIzaSyBCNskTKxgdbmwCh4BpVH0oo5-Xqt87MvY
```
